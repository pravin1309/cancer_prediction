```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("BreastCancer.csv")
df.head()
```

```python
for i,j in enumerate(df.columns):
    print(i,j)
```

```python
df.info()
```

```python
print(df['diagnosis'].unique())
print(f"Total Two Unique Value Are There:",df['diagnosis'].nunique())
print(df['diagnosis'].value_counts())
```

```python
# M = 0 = 212_Count
# B = 1 = 357_Count
df['diagnosis'] = df['diagnosis'].replace({'M': 0,"B": 1})
```

```python
df.drop(columns='id',axis=1,inplace=True)
df.drop(columns="Unnamed: 32",axis=1,inplace=True)
```

```python
df.sample(3)
```

```python
print(df.isna().sum())
sns.heatmap(df.isna())
```

```python
print("There Is No Value Are Duplicate:",df.duplicated().sum())
```

```python
df.info()
```

```python
# Id is not Important For Prediction 
plt.figure(figsize=(14,9))
sns.heatmap(df.corr(),annot=True,annot_kws={"size":6})
plt.show()
```

```python
for col in df.columns:
    if df[col].dtype != 'object' and col != 'diagnosis':
        plt.figure(figsize=(10,4))
        df.boxplot([col])
        plt.show()
```

```python
x = df.iloc[:,1:]
x
y = df["diagnosis"]
y
```

```python
# We are moving Without Trating OutLyers
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,stratify=y,random_state=100)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler,RobustScaler
```

```python
scalar=RobustScaler()
```

```python
x_train_scaled=scalar.fit_transform(x_train)
x_train_scaled
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_train_scaled
```

```python
x_test_scaled=scalar.transform(x_test)
x_test_scaled
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
x_test_scaled
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()
lg.fit(x_train,y_train)
```

```python
y_pred = lg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve
confusion_matrix(y_test,y_pred)
```

```python
accuracy= accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,y_pred))
```

```python
from sklearn import metrics
metrics.plot_roc_curve(lg,x_test,y_test)
plt.show()
```
